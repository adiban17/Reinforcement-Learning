```python
# importing libraries

import gymnasium

from huggingface_hub import notebook_login
from huggingface_sb3 import load_from_hub, package_to_hub

from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.monitor import Monitor
```

```python
import gymnasium as gym

# creating our environment
env = gym.make("LunarLander-v3")

# resetting the environment
observation, info = env.reset()

for _ in range(20):
  # ttake a random action
  action = env.action_space.sample()
  print("Action Taken:", action)

  # perform this action in the environment
  observation, reward, terminated, truncated, info = env.step(action)

  # if game is terminated
  if terminated or truncated:
    # reset the environment
    print("Environment is reset")
    observation, info = env.reset()

env.close()
```

```python
# creating a vectorized environment
env = make_vec_env("LunarLander-v3", n_envs=16)
```

```python
# using PPO to train our agent
model = PPO("MlpPolicy", env, verbose = 1)

# train the model
model.learn(total_timesteps=1000000)

# save the model
model_name = "ppo=LunarLander-v3"
model.save(model_name)
```

```python
# evaluating the agent
eval_env = Monitor(gym.make("LunarLander-v3", render_mode = "rgb_array"))
# evaluating the mean and standard reward
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic = True)
print(f"Mean Reward = {mean_reward} +/- {std_reward}")
```
